## When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
filtered_data=df_station[['install_dockcount','current_dockcount']].loc[~df_station['modification_date'].isnull()]
arr1=filtered_data['install_dockcount'].values
arr2=filtered_data['current_dockcount'].values
current_flag=0
install_flag=0

for i in arr1:
    for j in arr2:
        if i > j:
            install_flag+=1
            break
        else:
            current_flag+=1
            break
            
    # lose--> because the current flag which means # of bikes that in that station after modified is greater than the # of it before modification 
    
# test part
assert (install_flag > current_flag) == False
if install_flag >current_flag:
    print('gain')
else:
    print('lose')
```
